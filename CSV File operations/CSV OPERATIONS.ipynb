```python
import csv
with open( 'names.csv', 'r' ) as csv_file :
    csv_reader = csv.DictReader(csv_file)
    
    with open('new_file' ,'w') as new_file :
        
        fieldnames = ["first_name" , "last_name" , "email"] 
        csv_writer = csv.DictWriter(new_file , fieldnames= fieldnames , delimiter = '\t'  )
        
        
        # writeheader() method simply writes the first row of your csv file using the pre-specified fieldnames
        csv_writer.writeheader()    
        
        
        # writerows(): writerows method simply writes all the rows but in each row, it writes only the values(not keys).
        for line in csv_reader :
            #del line["email"]            # this removes the email index from file
            csv_writer.writerow(line )
        
        
        
```
